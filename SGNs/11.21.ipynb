```python
#get a list of the gene
gene_list=a['names'].tolist()
gene_list
```

```python
cell_type_markers={
'Schwann_cells': ['Sox10','Mpz','Pmp22'],
'Osteocytes':['Phex','Slc6a20a','Slc13a3'],
'Fibrocyte': ['Coch','Lum','Dcn'],
'Osteoblasts':['Dlx5','Runx2','Bglap'],
'chondrocyte':['Prg4','Slc26a2','Frzb'],
'Basal':['Cldn11', 'Clu','Nbl1'],
'immune':['Lyz2','Cd74','Ms4a6c','Ptprc'],
'endothelial':['Flt1','Robo4','Tie1'],
'Pericytes': ['Rgs5','Pdgfrb','Apoe'],
'Satellite_glial_cells':['Kcnj10','Slc12a2','Fbln2'],
'Oligo' :['Mobp','Mog','Mag','Olig1','Nkx2-2'],
'Type_2_neuron': ['Plk5','Prph','Rxrg'],
'Type_1A_neuron':['B3gat1','Mdga1','Obscn'],
'Type_1B_neuron':['Btbd9','Ttn','Rasgrp2'],
'Type_1C_neuron':['Pou4f1','Espn','Runx1','Lypd1'],
'nmSCs':['Scn7a','Cdh19','Apoe']

}
```

```python
#check if the genes is in the cell type markers
for i in gene_list:
    for key, value in cell_type_markers.items():
        if i in value:
            print(i,key)
```

```python
#check if the genes is in the cell type markers
for i in gene_list:
    for key, value in cell_type_markers.items():
        if i in value:
            print(i,key)
else:
    print("no")
```

```python
adata3=adata2.copy()
```

```python
adata3
```

```python
#remove those gene in gene_list from the adata2
adata3=adata3[:,~adata3.var_names.isin(gene_list)]
adata3
```

```python
adata3
```

```python
sc.tl.leiden(adata3,resolution=0.7)
sc.tl.umap(adata3,min_dist=0.5, spread=0.77)
sc.pl.umap(adata3,color=['leiden'],legend_loc='on data')
```

```python
sc.pl.umap(adata3,color=['cell_type'])
```

```python
sc.pp.highly_variable_genes(adata3,n_top_genes=1000,flavor='seurat_v3',layer='umi')
sc.pl.highly_variable_genes(adata3)
```

```python
adata3.X = adata3.layers['umi'].copy()
sc.pp.normalize_total(adata3, target_sum=None)
sc.pp.log1p(adata3)
adata3.layers['log1p'] = adata3.X.copy()
sc.pp.scale(adata3, max_value=10)
adata3.layers['Scale'] = adata3.X.copy()
adata3.X = adata3.layers['umi'].copy()
```

```python
sc.tl.pca(adata3, svd_solver='arpack')
sc.pl.pca_variance_ratio(adata3, log=True)
```

```python
sc.pp.neighbors(adata3, n_neighbors=50, n_pcs=13)
```

```python
sc.tl.leiden(adata3,resolution=0.7)
sc.tl.umap(adata3,min_dist=0.5, spread=0.77)
sc.pl.umap(adata3,color=['leiden'],legend_loc='on data')
```
