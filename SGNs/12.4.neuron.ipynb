```python
youth=youth[youth.obs['cell_type'].str.contains('neuron')]
sc.pl.umap(youth,color='cell_type')
```

```python
#combine youth and old
youth.obs['age']='youth'
old.obs['age']='old'
old.obs['sample_ID_Plate']=old.obs['platesID'].copy()
youth.X=youth.layers['umi'].copy()
old.X=old.layers['umi'].copy()
combined=youth.concatenate(old)
```

```python
sc.pp.highly_variable_genes(combined,n_top_genes=1500,flavor='seurat_v3')
sc.pp.normalize_per_cell(combined)
sc.pp.log1p(combined)
combined.layers['log1p']=combined.X.copy()
sc.pp.scale(combined,max_value=10)
combined.layers['scaled']=combined.X.copy()
sc.tl.pca(combined)
sc.pp.neighbors(combined,n_neighbors=15)
sc.tl.umap(combined)
sc.tl.leiden(combined,resolution=0.5)
sc.pl.umap(combined,color='leiden')
```

```python
sc.pl.umap(combined,color=['age','cell_type','leiden','Plk5','Prph','Rxrg','Kcnc2','Lypd1'])
```
