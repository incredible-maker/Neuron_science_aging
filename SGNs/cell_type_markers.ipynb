```python
# Cluster only the 'Osteoblasts' cells using the leiden algorithm
sc.tl.leiden(knwon_cluster_without_one_schwann_1, resolution=0.3, restrict_to=('cell_type', ['Osteoblasts']), key_added='leiden_Osteoblasts')

# Plot UMAP embeddings colored by the 'leiden_Osteoblasts' clusters
sc.pl.umap(knwon_cluster_without_one_schwann_1, color=['leiden_Osteoblasts'])
```

```python
sc.pl.umap(knwon_cluster_without_one_schwann_1,color=['cell_type','leiden','Slc47a1'],legend_loc='on data',title='cell_type',vmax='p95')
```
